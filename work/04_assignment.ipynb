```python
import pandas as pd
import matplotlib as plt

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
data = pd.read_csv('../data/state_facts.tsv', delimiter='\t')
```

```python
data.head()
```

```python
df = data[['city_1_pop','city_2_pop','city_3_pop','city_4_pop','city_5_pop']]
```

```python
plot = df.plot.bar(stacked=True,figsize=(10,10));
plot.set_xticklabels(data['State']);
plot.set_yticklabels(data['Pop_2020']);
```

```python
ax = plot = df.plot.bar(stacked=True,figsize=(10,10));
ax.set_xticklabels(data['State']);
data.plot(x='State', y='Pop_2020', ax=ax, ylabel="2020 State Population", title='Pop_Density Per City by State',figsize=(10,10));
```

```python
data[['city_1_pop','city_2_pop','city_3_pop','city_4_pop','city_5_pop']].plot.bar(stacked=True, legend=False,figsize=(10,10), colormap="twilight_shifted")
```
